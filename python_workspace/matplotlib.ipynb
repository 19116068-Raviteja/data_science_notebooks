Data Visualisation with Matplotlib

```python
import matplotlib.pyplot as plt
```

```python
x = [1,2,3,4,5]
y = [1,4,9,16,25]

## create a line plot
plt.xlabel('X axis')
plt.ylabel('Y label')
plt.title('Basic Line Plot')
plt.plot(x,y)
```

```python
x = [1,2,3,4,5]
y = [1,4,9,16,25]

## create a customised line plot
plt.xlabel('X axis')
plt.ylabel('Y label')
plt.title('customised line plot')
plt.plot(x,y,color='red',linestyle='--',marker='o',linewidth=2,markersize=9)
plt.grid(True)  # This enables the grid
```

```python
## multiple plots
## Sample data

x = [1,2,3,4,5]
y1 = [1,4,9,16,25]
y2 = [1,2,3,4,5]

plt.figure(figsize=(9,5))

plt.subplot(2,2,1) # 1 row and 2 cols and last is quarter
plt.plot(x,y1,color='green')
plt.title('Plot 1')

plt.subplot(2,2,2) 
plt.plot(y1,x,color='red')
plt.title('Plot 2')

plt.subplot(2,2,3) 
plt.plot(x,y2,color='blue')
plt.title('Plot 3')

plt.subplot(2,2,4)
plt.plot(x,y2,color='black')
plt.title('Plot 4')
```

```python
## bar plot

categories = ['A', 'B', 'C', 'D', 'E']
values = [5,7,3,8,6]

## create a bar plot
plt.xlabel('categories')
plt.ylabel('values')
plt.title('bar plot')

plt.bar(categories,values,color='purple')
 
```

#### Histograms

Histograms are used to represent the distribution of a dataset. They divide the data into bins and count the number of dara points in each bin.

```python
## sample data
data = [1,2,2,3,3,3,4,4,4,5,5,5,5,5]

plt.hist(data,bins=5,color='orange',edgecolor='black')
```

```python
## create a scatter plot 
# Sample data

x = [1,2,3,4,5]
y = [2,3,4,5,6]

plt.scatter(x,y,color='blue',marker='x')
```

```python
## pie chart

labels = ['A','B','C','D']
sizes = [30,20,40,10]

colors = ['green','purple','red','pink']
explode = (0.2,0,0,0) # move out the 1st slice

plt.pie(sizes,labels=labels,colors=colors,explode=explode,autopct="%1.1f%%",shadow=True)
```

```python
import pandas as pd
sales_data_df = pd.read_csv('sample.csv')
sales_data_df.head()
```

```python
## total sales by products
total_sales_by_products = sales_data_df.groupby('product')['sales'].sum()
total_sales_by_products
```

```python
total_sales_by_products.plot(kind='bar', color='teal')
```

```python
# sales over time
sales_data_df.groupby('date')['sales'].sum()

plt.plot(sales_data_df["date"],sales_data_df['sales'])
```
