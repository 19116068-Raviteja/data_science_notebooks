```python
## Handling missing values

df.isnull().any()
```

```python
df.isnull().sum()
```

```python
## Replace missing values with 0
df.fillna(0)
```

```python
## Replace missing values with mean of that column
df['sales_fillna']=df['sales'].fillna(df['sales'].mean())
df.head()
```

```python
df.dtypes
```

```python
## Renaming the columns
df = df.rename(columns={'date':'sales_date','category':'sales_category','value':'sales_value'})
df.head()
```

```python
## change datatypes
df['values_new'] = df['sales_value'].fillna(df['sales_value'].mean()).astype(int)
df.dtypes
```

```python
df['double_value'] = df['values_new'].apply(lambda x:x*2)
df.head()
```

```python
## Data aggregating and grouping

grouped_mean = df.groupby('product')['sales_value'].mean()
grouped_mean
```

```python
grouped_sum=df.groupby(['product','region'])['sales_value'].sum()
grouped_sum
```

```python
grouped_agg = df.groupby('region')['sales_value'].agg(['mean', 'sum', 'count'])
grouped_agg
```

```python
### merging and joining Dataframes
df1 = pd.DataFrame({'key': ['a','b','c'],'value1':[1,2,3]})
df2 = pd.DataFrame({'key': ['a','b','d'],'value2':[4,5,6]})
```

```python
df1
```

```python
df2
```

```python
## Merge DataFrame on the Key columns

pd.merge(df1,df2,on='key',how='inner')
```

```python
pd.merge(df1,df2,on='key',how='outer')
```

```python
pd.merge(df1,df2,on='key',how='left')
```

```python
pd.merge(df1,df2,on='key',how='right')
```
